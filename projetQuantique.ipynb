```
def init():
    backend = IBMQ.get_provider(hub='ibm-q').get_backend('ibmq_burlington')
    nqb = len(backend.properties().qubits)
    qc = QuantumCircuit(nqb,nqb)
    qc.h(range(nqb))
    qc.measure(range(nqb),range(nqb))
    return qc,backend
```

```
def exeQC(qc):
    #qiskit.tools.job_monitor(job_exp)
    counts = execute(qc, backend=backend, shots=1).result().get_counts(qc)
    return list(counts.keys())[0]
```

```
def randQC(qc, n=10):
    txt = ""
    x=0
    while x < n:
        #val = exeQC(qc)
        val = list(execute(qc,Aer.get_backend('qasm_simulator'),shots=1).result().get_counts(qc).keys())[0]
        txt += val
        x += len(val)
    return int(txt, 2)/(2**x)
```

```
def rand_seq_QC(qc, n=10):
    txt = ""
    x=0
    while x < n:
        val = exeQC(qc)
        txt += val
        x += len(val)
    return txt[:n]
```

```
def nrand(qc, n=100):
    L1 = []
    L2 = []
    for i in range(n):
        #quanticRand = randQC(qc)
        classicRand = random.random()
        #L1.append(quanticRand)
        L2.append(classicRand)
    return L1, L2
```

```
def piCalc(L):
    i = 0
    ret = 0.0
    while i < len(L)-1:
        if L[i]**2 + L[i+1]**2 < 1:
            ret += 1
        i += 2
    ret = ret*8/len(L)
    return ret
```

```
def test_statistique(seq):
  tot = 0
  for x in seq:
    tot += 2*int(x) -1 #On ajoute 1 lorsque c'est un 1, on retire 1 lorsque c'est un 0
  Sn = abs(tot)/math.sqrt(len(seq))
  Pvalue = math.erfc(Sn)
  if Pvalue < 0.01:
    #print(Pvalue," : Erreur, la séquence n'est pas aléatoire.")
    return False
  else:
    #print(Pvalue," : La séquence est bien aléatoire.")
    return True

test_statistique(seq_q[400:500])
test_statistique(seq_c[400:500])
```

```
def caracterisation_de_la_seq(seq,M):
  n = len(seq)
  nb_seq_non_valides = 0
  for i in range(n//M):
    if not(test_statistique(seq[M*i:M*(i+1)])):
      nb_seq_non_valides += 1
  print("Il y a ",nb_seq_non_valides," séquences de ",M," bits non valides dans la séquence à tester")

caracterisation_de_la_seq(seq_c,100)
caracterisation_de_la_seq(seq_q,100)
```
